```python
import torch
from transformers import AutoTokenizer, AutoModelWithLMHead
```

```python
tokenizer=AutoTokenizer.from_pretrained('T5-base')
model=AutoModelWithLMHead.from_pretrained('T5-base', return_dict=True)
```

```python
sequence = ("Flutterwave, Africa’s leading payments technology company, has secured $250 million in a Series D funding round led by Tiger Global and Avenir Growth. This latest investment values the company at $3 billion, solidifying its position as one of Africa’s most valuable startups.")
```

```python
inputs=tokenizer.encode("sumarize: " +sequence,return_tensors='pt', max_length=512, truncation=True)
```

```python
output = model.generate(inputs, min_length=80, max_length=100)
```

```python
summary=tokenizer.decode(output[0])
print(summary)
```
